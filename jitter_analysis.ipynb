# Jitter Analysis of Control Loop

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Experiment parameters

```python
sampling_time = 1000000 # ns
```

## Ubuntu 22.04

### Using chrono and while loop

```python
df1 = pd.read_csv('15-12-2022-16-17-57.csv')
```

```python
jitter1 = df1.diff().to_numpy() - sampling_time
```

Mean jitter

```python
np.mean(jitter1[1:])
```

Standard deviation of the jitter

```python
np.std(jitter1[1:])
```

```python
fig, ax = plt.subplots()

ax.plot(jitter1, marker='o', linestyle='None')

ax.set_xlabel('Samples')
ax.set_ylabel('Jitter [ns]')

fig.tight_layout()
```

```python
fig, ax = plt.subplots()

ax.plot(jitter1, marker='o', linestyle='None')

ax.set_xlabel('Samples')
ax.set_ylabel('Jitter [ns]')

ax.set_ylim(0, 400)

fig.tight_layout()

fig.savefig('jitter_ubuntu_22_04_1.pdf')
```
