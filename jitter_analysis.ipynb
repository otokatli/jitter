# Jitter Analysis of Control Loop

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('15-12-2022-08-36-23.csv')
```

```python
sampling_time = 1000 # ns
```

```python
jitter = df.diff().to_numpy() - sampling_time
```

Mean jitter

```python
np.mean(jitter[1:])
```

Standard deviation of the jitter

```python
np.std(jitter[1:])
```

```python
fig, ax = plt.subplots()

ax.plot(jitter, marker='o', linestyle='None')

ax.set_xlabel('Samples')
ax.set_ylabel('Jitter [ns]')

fig.tight_layout()
```

```python
fig, ax = plt.subplots()

ax.plot(jitter, marker='o', linestyle='None')

ax.set_xlabel('Samples')
ax.set_ylabel('Jitter [ns]')

ax.set_ylim(0, 300)

fig.tight_layout()

fig.savefig('jitter_ubuntu_22_04.pdf')
```
